```python
from PIL import Image
from IPython.display import display

directory = "data/vcl_50"
```

```python
image = Image.open(f"{directory}/9/coreset_mle_all_w2.0.png")
image
```

```python
images = []

for i in range(10):
    image = Image.open(f"{directory}/{i}/coreset_mle_all_w2.0.png")
    for j in range(i+1):
        cropped = image.crop((30*j+2, 2, 30*j+30, 30))
        images.append(cropped)
```

```python
images = []

for i in range(10):
    image = Image.open(f"{directory}/{i}/image_ep49_w2.0.png")
    for j in range(i+1):
        cropped = image.crop((30*j+2, 30*3+2, 30*j+30, 30*3+30))
        images.append(cropped)
```

```python
grid_width = 28 * 10
grid_height = 28 * 10
grid_image = Image.new("RGB", (grid_width, grid_height), "white")

# Paste the images into the grid
row_start = 0
image_index = 0
for row in range(1, 10+1):
    for col in range(row):
        x = col * 28
        y = row_start
        grid_image.paste(images[image_index], (x, y))
        image_index += 1
    row_start += 28

display(grid_image)
```

```python
grid_image.save("plots/vcl_50.png")
```

```python
import cv2

for i in [50, 200, 500]:
    img = cv2.imread(f"plots/vcl_{i}.png")
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Invert the colors of the grayscale image
    inverted_image = cv2.bitwise_not(gray_image)
    cv2.imwrite(f"plots/vcl_{i}_inv.png", inverted_image)
```
